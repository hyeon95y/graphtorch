```python
import copy


def show_graph(graph):
    def show_without_nan(matrix):
        display(copy.deepcopy(matrix).fillna("_"))
        return

    show_without_nan(graph.adjacency_matrix)
    for feature in graph.feature_matrix.keys():
        print(feature)
        show_without_nan(graph.feature_matrix[feature])
    return
```

# Create your graph representing network

```python
import graphtorch as gt

in_dim = 2
hidden_node_dims = [2, 4, 3]
out_dim = 3
graph = gt.Graph(in_dim, out_dim, hidden_node_dims)

show_graph(graph)
```

# Split input, output layer to individual nodes

```python
in_dim = 2
hidden_node_dims = [2, 4, 3]
out_dim = 3
graph = gt.Graph(
    in_dim, out_dim, hidden_node_dims, split_input_layer=True, split_output_layer=True
)

show_graph(graph)
```

# Wire network by yourself

```python
in_dim = 2
hidden_node_dims = [2, 4, 3]
out_dim = 3
graph = gt.Graph(in_dim, out_dim, hidden_node_dims)

graph["I:0", "H:0"] = True
graph["H:0", "H:1"] = True
graph["H:1", "H:2"] = True
graph["H:2", "O:0"] = True

show_graph(graph)
```

# Generate network from your graph

```python
layer = gt.GraphLayer(graph)
layer.wires
```
